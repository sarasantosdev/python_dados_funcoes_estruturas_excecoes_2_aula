# Funções - Aula

```python
# Notas do estudante

notas = {"1º Trimestre": 8.5, "2º Trimestre": 9.5, "3º Trimestre": 7}
notas
```

```python
# calculando a soma sem utilizar a função embutida sum()

soma = 0
for nota in notas.values():
  soma += nota

media = soma / len(notas.values())
round(media, 1)
```

```python
# calculando a soma utilizando a função embutida sum()

soma = sum(notas.values())

media = soma / len(notas.values())
round(media, 1)
```

## Função com parâmetro

```python
def media(nota_1, nota_2, nota_3):
  calculo = (nota_1 + nota_2 + nota_3) / 3
  print(calculo)
```

```python
media(3,6,9)
```

Os parâmetros são declarados no momento da criação da função, enquanto que os argumentos são os valores desses parâmetros e serão passados ao chamarmos a função.

```python
nota1 = 8
nota2 = 9
nota3 = 7

media(nota_1=nota1, nota_2=nota2, nota_3=nota3)
```

```python
# Notas do estudante
notas = [6.0, 7.0, 9.0, 5.0]
```

```python
# minha resolução
def boletim(lista):
  media = sum(lista) / len(lista)

  if media >= 6:
    print("Aprovado(a)!")
  else:
    print("Reprovado(a)!")

  return media
```

```python
boletim(notas)
```

```python
# resolução do professor
def boletim(lista):
  media = sum(lista) / len(lista)

  if media >= 6:
    situacao = "Aprovado(a)"
  else:
    situacao = "Reprovado(a)"

  return (media, situacao)
```

```python
boletim(notas)
```

```python
# separando os valores de tupla ou lista em duas variáveis
media, situacao = boletim(notas)
```

```python
print(f'O(a) estudante atingiu uma média de {media} e foi {situacao}')
```

## Função Lambda

São funções anônimas úteis para quando não precisaremos reutilizar a função

```python
nota1 = float(input('Informe a 1ª nota: '))
nota2 = float(input('Informe a 2ª nota: '))
nota3 = float(input('Informe a 3ª nota: '))
```

```python
# minha resolução
media = lambda nota1, nota2, nota3: (nota1*3 + nota2 *2 + nota3*5)/(3+2+5)
media(nota1, nota2, nota3)
```

## Função map

Mapeia valores

```python
# Notas do(a) estudante
notas = [6.0, 7.0, 9.0, 5.5, 8.0]
qualitativo = 0.5
```

```python
notas_finais = list(map(lambda x: x + qualitativo, notas ))
notas_finais
```

```python
# mesmo resultado utilizando compreensão de lista
n = [x + qualitativo for x in notas]
n
```
